```python
import pandas as pd
df = pd.read_csv("/content/ABCTargetlatlong.csv")
df.head()
```

```python
import folium
m = folium.Map(location = [df.Latitude.mean(),df.Longitude.mean()], zoom_start=10, tiles = None)
folium.TileLayer(tiles = "openstreetmap", control=False).add_to(m)
colors = ["red", "blue"]
for index, store in enumerate(df.Store.unique()):
  temp = df[df.Store == store]
  fg = folium.FeatureGroup(name = store)
  for i in range(len(temp)):
    folium.Marker(location = [temp.Latitude.iloc[i],temp.Longitude.iloc[i]],
                  icon = folium.Icon(color = colors[index]),
                  popup = "<strong>Store Type: </strong>" + temp.Store.iloc[i] + "<br><strong>Address: </strong>" + temp.Address.iloc[i] + "<br><strong>City: </strong>" + temp.City.iloc[i]).add_to(fg)
  fg.add_to(m)
folium.LayerControl(collapsed = False).add_to(m)
m
```

```python
from google.colab import files
m.save("ABC_Target.html")
files.download("ABC_Target.html")
```
